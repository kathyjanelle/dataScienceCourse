### Exercise 1. 

Read Total Profit for all months and show it using a line plot with Month Number as your baseline

Total profit data provided for each month. Generated line plot must include the following properties:

1. X label name = Month Number
2. Y label name = Total profit

````python
# Exercise 1
import pandas as pd 
import matplotlib.pyplot as plt

# Extract the data we're interested in
cosmetics = pd.read_csv('./data/company_sales_data.csv')
month, profit = cosmetics['month_number'], cosmetics['total_profit']

#plt.axis(aspect='equal')
plt.plot(month, profit)
plt.grid()
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Total Profit of All months')
````

### Exercise 2. 

Get Total Units Sold of all months and show line plot with the following Style properties

Generated line plot must include following Style properties:

1. Line Style dotted and Line-color should be red
2. Show legend at the lower right location.
3. X label name = Month Number
4. Y label name = Units Sold Number
5. Add a circle marker.
6. Line marker color as read
7. Line width should be 3

````python
# Exercise 2
import pandas as pd 
import matplotlib.pyplot as plt

# Extract the data we're interested in
cosmetics = pd.read_csv('./data/company_sales_data.csv')
month, profit = cosmetics['month_number'], cosmetics['total_profit']

#To plot the graph with grid, title, legend  
plt.plot(month, profit, label = 'Profit based on months', color = 'r', marker = 'o', markerfacecolor = 'r', linestyle = 'dotted', linewidth = 3)
plt.grid()
plt.xlabel('Month Number')
plt.ylabel('Units Sold Number')
plt.legend(loc = 'lower right',framealpha=1, frameon=True);
plt.title('Total Profit of All months')
````

### Exercise 3.

Read all product sales data and show it using a multiline plot

Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).

````python
# facecream 	facewash 	toothpaste 	bathingsoap 	shampoo 	moisturizer 	
import pandas as pd 
import matplotlib.pyplot as plt

# Extract the data we're interested in
cosmetics = pd.read_csv('./data/company_sales_data.csv')
month = cosmetics['month_number']
faceCreamData = cosmetics['facecream']
faceWashData = cosmetics['facewash']
toothPasteData = cosmetics['toothpaste']
bathingSoapData = cosmetics['bathingsoap']
shampooData = cosmetics['shampoo']
moisturizerData = cosmetics['moisturizer']


#To plot the graph with grid, title, legend  
plt.plot(month, faceCreamData, label = 'facecream', color = 'k', linestyle = 'dotted', linewidth = 3)
plt.plot(month, faceWashData, label = 'facewash', color = 'b', linestyle = '--', linewidth = 3)
plt.plot(month, toothPasteData, label = 'toothpaste', color = 'r', linestyle = '-', linewidth = 3)
plt.plot(month, bathingSoapData, label = 'bathingsoap', color = 'g', linestyle = 'dashdot', linewidth = 3)
plt.plot(month, shampooData, label = 'shampoo', color = 'c', marker = 'o', markerfacecolor = 'c', linestyle = '-', linewidth = 3)
plt.plot(month, moisturizerData, label = 'moisturizer', color = 'y', marker = '', markerfacecolor = 'y', linestyle = 'dotted', linewidth = 3)

plt.grid()
plt.xlabel('Month Number')
plt.ylabel('Units Sold Number')
plt.legend(loc = 'upper left',framealpha=1, frameon=True);
plt.title('All Product Sales')
````

### Exercise 4.

Read toothpaste sales data of each month and show it using a scatter plot

Also, add a grid in the plot. gridline style should “–“.

````python
import pandas as pd 
import matplotlib.pyplot as plt

# Extract the data we're interested in
cosmetics = pd.read_csv('./data/company_sales_data.csv')
month,toothPasteSalesData = cosmetics['month_number'], cosmetics['toothpaste']

#plot scatter plot 
plt.scatter(month,toothPasteSalesData)
plt.xlabel('Month Number')
plt.ylabel('Toothpaste Sales Data')
plt.title('Toothpaste sales data')
plt.grid()
````

### Exercise 5.

Read face cream and face wash product sales data and show it using the bar chart

The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

````python
import pandas as pd 
import matplotlib.pyplot as plt

# Extract the data we're interested in
cosmetics = pd.read_csv('./data/company_sales_data.csv')
month = cosmetics['month_number'] 
faceWashSales = cosmetics['facewash']
faceCreamSales = cosmetics['facecream']

#Display a bar graph
plt.bar(month,faceWashSales, width = 0.25, label = 'Facewash')
plt.bar(month,faceCreamSales, width = -0.25, label = 'Facecream')
plt.xlabel('Month Number')
plt.ylabel('Number of Units Sold')
plt.legend(loc = 'upper left', framealpha=1, frameon=True)
plt.title('Facecream and Facewash Product Sales')
````
